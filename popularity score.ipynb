```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
csv_file_path = 'path/to/your/Final Content Data set.csv'
data = pd.read_csv(r'D:\Code_Box\Social Buzz\Final Content Data set.csv')

# Calculate the aggregate popularity score for each category
aggregate_popularity = data.groupby('Category')['Score'].sum().sort_values(ascending=False)

# Plot the bar chart for aggregate popularity score by category
plt.figure(figsize=(12, 6))
aggregate_popularity.plot(kind='bar', color='Purple')
plt.title('Aggregate Popularity Score by Category')
plt.xlabel('Category')
plt.ylabel('Aggregate Popularity Score')
plt.xticks(rotation=45)
plt.show()
```
